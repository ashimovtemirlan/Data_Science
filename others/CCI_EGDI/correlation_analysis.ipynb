```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading and wrangling data

```python
# Reading data
data = pd.read_excel('data.xlsx')
# Data wrangling
data['Time Period'] = pd.to_datetime(data['Time Period'].astype(str) + '-01-01')
```

# Analysis

```python
# All columns and countries
columns = data.columns
countries = ['Estonia','Lithuania','Kazakhstan','Russia','Belarus',
             'Latvia','Georgia','Armenia','Ukraine','Azerbaijan',
             'Moldova','Kyrgyzstan','Uzbekistan','Tajikistan','Turkmenistan']

# Graph
f, ax = plt.subplots(5, 3, figsize=(40,40))
for i, country in enumerate(countries):
    columns_for_corr = []
    for column in [column for column in columns if country in column]:
        columns_for_corr.append(column)
        sns.lineplot(x='Time Period', y=column, data=data, label=column, ax=ax[i%5][i//5])
        ax[i%5][i//5].legend()
        ax[i%5][i//5].grid(True)
        ax[i%5][i//5].set_ylabel('')
    ax[i%5][i//5].set_title(f'Correlation={round(data[columns_for_corr].corr().iloc[0,1], 2)}')
f.savefig('results/corr.png', dpi=300)
plt.show()
```

```python
# Total Correlation
data.corr().to_excel('results/corr.xlsx', index=True)

data.corr()
```
