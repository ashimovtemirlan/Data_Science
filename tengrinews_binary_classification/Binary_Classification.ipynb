# Binary classification of "Tengrinews" news
In this project I want to collect news from tengrinews.kz and train binary classification model (Logistic function) on this data (All news are in russian language). There are classes: "Экономика" (Economics), "Спорт" (Sport).

# Data collection



```
import urllib.request
from bs4 import BeautifulSoup

def get_html(url):
  response = urllib.request.urlopen(url)
  return response.read()

def parse_news(html):
  soup = BeautifulSoup(html)
  try:
    h1 = soup.find('h1', class_='tn-content-title')
    div = soup.find('div', class_='tn-news-text')
    news = h1.text.strip().split()
    for p in div.find_all('p'):
      news += p.text.split()
    return news
  except:
    None

def parse_links(html):
  soup = BeautifulSoup(html)
  list_news = []
  div = soup.find('div', class_='tn-article-grid')
  for conference in div.find_all('div', class_='tn-article-item'):
    for href in conference.find_all('a', href_=''):
      list_news.append(parse_news(get_html('https://tengrinews.kz' + href['href'])))
  return list_news
```

```
data_economics = []
data_sports = []
```

```
def economics(n):
  global data_economics
  for i in range(2, n + 2):
    data_economics += parse_links(get_html("https://tengrinews.kz/economic/page/" + str(i) + "/"))

def sports(n):
  global data_sports
  for i in range(2, n + 2):
    data_sports += parse_links(get_html("https://tengrinews.kz/sports/page/" + str(i) + "/"))
```

```
n = int(input("Amount of pages to parse:"))

economics(n)
sports(n)
```

```
import pymorphy2
from string import punctuation

morph = pymorphy2.MorphAnalyzer()
no_need = ['PREP', 'PRED', 'CONJ', 'PRCL', 'INTJ', 'PNCT', 'ROMN', 'NUMB', 'UNKN', 'ADVB', 'NPRO']

data_economics_clear = []
data_sports_clear = []

for i, news in enumerate(data_economics):
  data_economics_clear.append([])
  for word in news:
    word_1 = word.strip(punctuation)
    p = morph.parse(word_1)[0]
    if str(p.tag)[0:4] not in no_need:
      data_economics_clear[i].append(p.normal_form)

for i, news in enumerate(data_sports):
  data_sports_clear.append([])
  for word in news:
    word_1 = word.strip(punctuation)
    p = morph.parse(word_1)[0]
    if str(p.tag)[0:4] not in no_need:
      data_sports_clear[i].append(p.normal_form)
```

```
i = 0
for news in data_economics_clear:
  i += len(news)
for news in data_sports_clear:
  i += len(news)
```

```
print('AFTER CLEANING! \n')
print('Total words amount     - ' + str(i) + '.')
print('Economics news amount  - ' + str(len(data_economics_clear)) + '.', 'News :', data_economics_clear)
print('Sports news amount     - ' + str(len(data_sports_clear)) + '.', 'News :', data_sports_clear)
```

```
dictionary = []

for news in data_economics_clear:
  for word in news:
    if word not in dictionary:
      dictionary.append(word)

for news in data_sports_clear:
  for word in news:
    if word not in dictionary:
      dictionary.append(word)

print('Dictionary size - ' + str(len(dictionary)) + '.')
```

```
dictionary_data = {'Класс':[]}

for word in dictionary:
  dictionary_data[word] = []

for news in data_economics_clear:
  dictionary_data['Класс'] += ['Экономика']
  for word in dictionary:
    dictionary_data[word] += [news.count(word)]

for news in data_sports_clear:
  dictionary_data['Класс'] += ['Спорт']
  for word in dictionary:
    dictionary_data[word] += [news.count(word)]
```

```
import pandas as pd

pd_data = pd.DataFrame.from_dict(dictionary_data)
```

# Data analysis

```
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, shuffle = True, random_state = 1)
```

```
X_train = X_train.T
X_test = X_test.T
Y_train = Y_train.reshape((1, Y_train.shape[0]))
Y_test = Y_test.reshape((1, Y_test.shape[0]))
```

```
print('Shape of the training set X: ', X_train.shape)
print('Shape of the test set X:     ', X_test.shape)
print('Shape of the training set Y: ', Y_train.shape)
print('Shape of the test set Y:     ', Y_test.shape)
```

```
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
```

```
def initialize_with_zeros(n):
  w = np.zeros((n, 1))
  b = 0
  return w, b  
```

```
def propagate(w, b, X, Y):
  m = Y.shape[1]

  ## Forward Propogation ##
  A = sigmoid( np.dot(w.T, X) + b )
  cost = -(1/m)*np.sum( Y*np.log(A) + (1-Y)*np.log(1-A) )

  ## Backward Propogation ##
  dw = (1/m)*np.dot(X, (A-Y).T)
  db = (1/m)*np.sum(A-Y)

  grads = {'dw': dw, 
           'db': db}

  return grads, cost
```

```
def optimize(w, b, X, Y, num_iterations, learning_rate):

  for i in range(num_iterations):
    grads, cost = propagate(w, b, X, Y)
    
    dw = grads['dw']
    db = grads['db']

    w = w - learning_rate*dw
    b = b - learning_rate*db

    if i % 100 == 0:
      print('Cost after '+str(i)+' iterations:', cost)

    params = {'w': w,
              'b': b}

  return params
```

```
def predict(w, b, X):
  m = X.shape[1]
  Y_prediction = np.zeros((1, m))
  w = w.reshape((X.shape[0], 1))

  A = sigmoid( np.dot(w.T, X) + b )

  for i in range(m):
    if A[0, i] >= 0.5:
      Y_prediction[0, i] = 1
    else:
      Y_prediction[0, i] = 0

  return Y_prediction
```

```
def model(X_train, X_test, Y_train, Y_test, num_iterations, learning_rate):
  w, b = initialize_with_zeros(X_train.shape[0])
  params = optimize(w, b, X_train, Y_train, num_iterations, learning_rate)

  w = params['w']
  b = params['b']

  Y_prediction_train = predict(w, b, X_train)
  Y_prediction_test = predict(w, b, X_test)

  print('Train accuracy: '+str( 100*(1 - np.mean(np.abs(Y_prediction_train - Y_train))) )+'%')
  print('Test accuracy:  '+str( 100*(1 - np.mean(np.abs(Y_prediction_test - Y_test))) )+'%')    
```

```
model(X_train, X_test, Y_train, Y_test, num_iterations = 2000, learning_rate = 0.005)
```
