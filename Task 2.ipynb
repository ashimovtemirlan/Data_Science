### - Datetime переменные: 'DATE_FRM_LST_KASPIKZ_PMT', 'DATE_LST_PKB_RQST'.

```python
def transform_month(x):
    month = {'JAN': '01', 'FEB': '02', 'MAR': '03', 'APR': '04', 'MAY': '05', 'JUN': '06',
             'JUL': '07', 'AUG': '08', 'SEP': '09', 'OCT': '10', 'NOV': '11', 'DEC': '12'}
    return month[x]

def transform_date(x):
    try:
        return x[5:9] + '-' + transform_month(x[2:5]) + '-' + x[:2]
    except:
        return np.nan
```

```python
pd_train_3 = pd_train_2.copy()
pd_validate_3 = pd_validate_2.copy()

pd_train_3['DATE_FRM_LST_KASPIKZ_PMT'] = pd.to_datetime( pd_train_2['DATE_FRM_LST_KASPIKZ_PMT'].apply(lambda x: transform_date(x)) )
pd_train_3['DATE_LST_PKB_RQST'] = pd.to_datetime( pd_train_2['DATE_LST_PKB_RQST'].apply(lambda x: transform_date(x)) )

pd_validate_3['DATE_FRM_LST_KASPIKZ_PMT'] = pd.to_datetime( pd_validate_2['DATE_FRM_LST_KASPIKZ_PMT'].apply(lambda x: transform_date(x)) )
pd_validate_3['DATE_LST_PKB_RQST'] = pd.to_datetime( pd_validate_2['DATE_LST_PKB_RQST'].apply(lambda x: transform_date(x)) )
```

### - Несбалансированная база данных.

```python
Y.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state = 0)
X_resampled, Y_resampled = ros.fit_resample(X, Y)
```

### - Data imputation.

```python
from sklearn.impute import SimpleImputer

imp_mean = SimpleImputer(strategy = 'mean')
imp_mean.fit(X)
```

```python
X_final = imp_mean.transform(X_resampled)
X_validate_final = imp_mean.transform(X_validate)
```

### Разбиение на train и test части.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_final, Y_resampled, train_size = 0.9, shuffle = True, random_state = 1)
```

# Machine Learning.

### - RandomForestClassifier.

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(criterion = 'entropy', max_features = 'log2')
clf.fit(x_train, y_train)
```

```python
y_pred_train = clf.predict(x_train) 
y_pred_test = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print('Confusion matrix для train датасета:')
print(confusion_matrix(y_train, y_pred_train))
print('')
print('Classification_report для test датасета:')
print(classification_report(y_train, y_pred_train))
```

```python
print('Confusion matrix для train датасета:')
print(confusion_matrix(y_test, y_pred_test))
print('')
print('Classification_report для test датасета:')
print(classification_report(y_test, y_pred_test))
```
